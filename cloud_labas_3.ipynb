# E-commerce Sales Analysis

Part I: Randomized Sales Data Generation (NumPy)

```python
import numpy as np
import pandas as pd
```

```python
rollno = 8025320088
np.random.seed(rollno %(2**32 - 1) )
```

```python
salesdata = np.random.randint(1000, 5000, size=(12, 4))
salesdata
```

```python
columns = ["Electronics", "Clothing", "Home & Kitchen", "Sports"]
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun",
          "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]

df = pd.DataFrame(salesdata, columns=columns, index=months)
df
```

Part II: Data Manipulation and Analysis (Pandas)

```python
df.head()
```

```python
df.describe()
```

```python
categorytot = df.sum()
categorytot
```

```python
monthlytot = df.sum(axis=1)
monthlytot
```

```python
growth = df.pct_change()

avg_growth = growth.mean()
avg_growth
```

```python
df["Total Sales"] = df.sum(axis=1)
df["Growth Rate"] = df["Total Sales"].pct_change() * 100
df
```

```python
if rollno % 2 == 0:
    df["Electronics"] = df["Electronics"] * 0.9
else:
    df["Clothing"] = df["Clothing"] * 0.85
df
```

Part III: Visualizations (Matplotlib & Seaborn)

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(10, 6))

for col in columns:
    plt.plot(df.index, df[col], marker='o', label=col)

plt.title("Monthly Sales Trends")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.legend()
plt.grid()

plt.show()
```

```python
plt.figure(figsize=(8, 6))

sns.boxplot(data=df[columns])

plt.title("Sales Distribution by Category")
plt.ylabel("Sales")

plt.show()
```

# Q.2 For the array: array = np.array([[1, -2, 3],[-4, 5, -6]])

```python
import numpy as np
array = np.array([[1, -2, 3],[-4, 5, -6]])
```

i. Find element-wise absolute value

```python
abs_array = np.abs(array)
abs_array
```

ii. Percentiles (25th, 50th, 75th)

```python
flat = array.flatten()
percenflat = np.percentile(flat, [25, 50, 75])
percenflat
```

```python
percencol = np.percentile(array, [25, 50, 75], axis=0)
percencol
```

```python
percenrow = np.percentile(array, [25, 50, 75], axis=1)
percenrow
```

iii. Mean, Median, Standard Deviation

```python
meanflat = np.mean(flat)
print(meanflat)

medianflat = np.median(flat)
print(medianflat)

stdflat = np.std(flat)
print(stdflat)
```

```python
meancol = np.mean(array, axis=0)
print(meancol)

mediancol = np.median(array, axis=0)
print(mediancol)

stdcol = np.std(array, axis=0)
print(stdcol)
```

```python
meanrow = np.mean(array, axis=1)
print(meanrow)

medianrow = np.median(array, axis=1)
print(medianrow)

stdrow = np.std(array, axis=1)
print(stdrow)
```

# Q.4 WAP to swap two elements in a list in Python Using a Temporary Variable.

```python
list1 = [20, 40, 60, 80, 100]

i = 1
j = 3

print("Original List:", list1)

temp = list1[i]
list1[i] = list1[j]
list1[j] = temp

print("List after swapping:", list1)
```

# Q.5 Write a Python program to swap two elements in a set by converting it to a list first.

```python
s = {20, 50, 80, 110}
print("Original Set:", s)

list2 = list(s)

i = 0
j = 2

temp = list2[i]
list2[i] = list2[j]
list2[j] = temp

s = set(list2)
print("Set after swapping:", s)
```
